```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equation coefficients
a, b, c, d = 1, 2, 3, 4

# Create grid points for x and y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Calculate z for the plane equation z = ax + by + d
z = (-a * x - b * y - d) / c

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import plotly.graph_objs as go

# Define the plane equation coefficients
a, b, c, d = 1, 2, 3, 4

# Create grid points for x and y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Calculate z for the plane equation z = ax + by + d
z = (-a * x - b * y - d) / c

# Create a surface trace for the plane
plane = go.Surface(x=x, y=y, z=z, opacity=0.5)

# Create the figure
fig = go.Figure(data=[plane])

# Update layout
fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
    )
)

# Show the plot
# fig.show()
```
